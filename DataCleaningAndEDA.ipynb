```python
import pandas as pd
```

```python
og_data = pd.read_csv("a1_data.csv")
```

```python
og_data.head()
```

```python
#data grouped & counted by crime type
crim_d = og_data.groupby('Primary Type').count()[["ID"]]
crim_d
crim_d.to_csv("byPrimaryType.csv")
```

```python
crim_d2 = og_data.groupby('Description').count()[["ID"]].sort_values(["ID"], ascending=False)
crim_d2.to_csv("byDescription.csv")
```

```python
crim_d3 = og_data.groupby('Location Description').count()[["ID"]]
crim_d3.to_csv("byLocationType.csv")
```
